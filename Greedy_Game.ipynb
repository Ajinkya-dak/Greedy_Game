# **Question 1**

```python
import pandas as pd
import datetime

# read the user signup data
user_signup = pd.read_csv('user_signup.csv')

# read the user offer completion data
user_offer_completion = pd.read_csv('user_offer_completion.csv')

# read the rewards details
rewards = pd.read_csv('rewards.csv')

# merge user signup and user offer completion data
merged_data = pd.merge(user_signup, user_offer_completion, on='user_id', how='left')

# merge the result with rewards data
merged_data = pd.merge(merged_data, rewards, on='reward_id', how='left')

# calculate total revenue generated by the app for each user
revenue_per_user = merged_data.groupby('user_id')['total_revenue_in_paise'].sum()

# calculate the number of users acquired through each marketing channel
users_per_channel = user_signup.groupby('utm_source')['user_id'].nunique()

# calculate the total revenue generated for each marketing channel
revenue_per_channel = merged_data.groupby('utm_source')['total_revenue_in_paise'].sum()

# calculate the average revenue per user (ARPU) for each marketing channel
arpu_per_channel = revenue_per_channel / users_per_channel

# calculate the average customer lifespan
current_date = datetime.datetime.now()
user_signup['created_at'] = pd.to_datetime(user_signup['created_at'])
user_signup['lifespan'] = (current_date - user_signup['created_at']).dt.days

# calculate the customer lifetime value (LTV) for each marketing channel
ltv_per_channel = arpu_per_channel * user_signup.groupby('utm_source')['lifespan'].mean()

print(ltv_per_channel)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load the Sikka_numbers table
sikka_numbers = pd.read_csv(path5)

# Filter the data to include only the relevant columns and October data
sikka_october = sikka_numbers[['Date', 'DAU', 'Referrals']].loc[sikka_numbers['Date'].str.contains('2022-10')]

# Create a linear regression model
model = LinearRegression()
model.fit(sikka_october[['DAU']], sikka_october[['Referrals']])

# Use the model to predict the number of referrals for each day of the first 15 days of November
sikka_november = sikka_numbers[['Date', 'DAU']].loc[sikka_numbers['Date'].str.contains('2022-11')]
sikka_november['predicted_referrals'] = model.predict(sikka_november[['DAU']])

# Print the predicted number of referrals for each day of the first 15 days of November
print("Predicted number of referrals for each day of the first 15 days of November:")
print(sikka_november[['Date', 'predicted_referrals']])
```

```python
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv(path6)

# Calculate the modified Z-score for each metric
med = data.median()
mad = (data - med).abs().median()
modified_z_score = 0.6745 * (data - med) / mad

# Calculate the threshold for each metric
thresholds = np.arange(0, 5, 0.1)
best_threshold = {}
for col in data.columns:
    scores = []
    for t in thresholds:
        s = modified_z_score[col].apply(lambda x: abs(x) > t).sum()
        scores.append(s)
    best_t = thresholds[np.argmin(scores)]
    best_threshold[col] = best_t

# Print the best threshold for each metric
print(best_threshold)
```

```python
# Calculate the modified Z-score for the 'ae_revenue' metric
med = data['ae_revenue'].median()
mad = (data['ae_revenue'] - med).abs().median()
modified_z_score = 0.6745 * (data['ae_revenue'] - med) / mad

# Find the data points where the modified Z-score exceeds the threshold
threshold = 4.9
anomalies = data[modified_z_score.abs() > threshold]

# Group the anomalies by 'app_id'
anomalies_by_app = anomalies.groupby('app_id').size()

# Print the anomalies by 'app_id'
print(anomalies_by_app)
```
